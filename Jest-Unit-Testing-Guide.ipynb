# Add Unit Testing with Jest

This guide shows how to add unit testing with Jest to your Pages, Button, and Input components in your Vite + React project.

## 1. Set Up Jest in the Project

Install Jest, React Testing Library, and related dependencies:

```sh
npm install --save-dev jest @testing-library/react @testing-library/jest-dom @testing-library/user-event ts-jest
```

Add the following to your `package.json` scripts:
```json
"scripts": {
  "test": "jest"
}
```

Create a `jest.config.js` file:
```js
module.exports = {
  preset: 'ts-jest',
  testEnvironment: 'jsdom',
  moduleNameMapper: {
    "\\.(css|less|scss|sass)$": "identity-obj-proxy"
  }
};
```

Optionally, add a `setupTests.ts` file for custom setup (e.g., jest-dom):
```ts
import '@testing-library/jest-dom';
```
And reference it in your Jest config:
```js
setupFilesAfterEnv: ['<rootDir>/src/setupTests.ts'],
```

## 2. Write Unit Tests for Pages

Create a `__tests__` folder or `*.test.tsx` files in your `src/pages` directory. Example for `Home` page:

```tsx
// src/pages/__tests__/Home.test.tsx
import { render, screen } from '@testing-library/react';
import Home from '../Home';

describe('Home Page', () => {
  it('renders the hero section', () => {
    render(<Home />);
    expect(screen.getByText(/Finestar/i)).toBeInTheDocument();
  });
});
```

Repeat for other pages (e.g., CartPage, CheckoutPage) to verify rendering and key UI elements.

## 3. Write Unit Tests for Button Component

Create a test file for the Button component:

```tsx
// src/components/__tests__/Button.test.tsx
import { render, screen, fireEvent } from '@testing-library/react';
import Button from '../Button';

describe('Button', () => {
  it('renders with children', () => {
    render(<Button>Click Me</Button>);
    expect(screen.getByText('Click Me')).toBeInTheDocument();
  });

  it('calls onClick when clicked', () => {
    const handleClick = jest.fn();
    render(<Button onClick={handleClick}>Click</Button>);
    fireEvent.click(screen.getByText('Click'));
    expect(handleClick).toHaveBeenCalledTimes(1);
  });

  it('shows loading state', () => {
    render(<Button loading>Loading...</Button>);
    expect(screen.getByText('Loading...')).toBeDisabled();
  });
});
```

## 4. Write Unit Tests for Input Component

Create a test file for the Input component:

```tsx
// src/components/__tests__/Input.test.tsx
import { render, screen, fireEvent } from '@testing-library/react';
import Input from '../Input';

describe('Input', () => {
  it('renders with placeholder', () => {
    render(<Input placeholder="Type here" />);
    expect(screen.getByPlaceholderText('Type here')).toBeInTheDocument();
  });

  it('calls onChange when value changes', () => {
    const handleChange = jest.fn();
    render(<Input onChange={handleChange} />);
    fireEvent.change(screen.getByRole('textbox'), { target: { value: 'abc' } });
    expect(handleChange).toHaveBeenCalled();
  });

  it('shows the correct value', () => {
    render(<Input value="hello" readOnly />);
    expect(screen.getByDisplayValue('hello')).toBeInTheDocument();
  });
});
```

## 5. Run Tests and View Results

Run all tests using the command:

```sh
npm test
```

Or, for watch mode:
```sh
npm test -- --watch
```

You can also use the VS Code Jest extension for an integrated test runner and results pane.

Test output will show which tests passed or failed, and coverage if configured.
